**Задание 1**

Импортировать данные о ежемесячных ценах на нефть марки Brent по ссылке https://datahub.io/core/oil-prices/r/brent-month.csv Подсчитать средние цены на нефть, минимальное и максимальное значение, среднеквадратическое отклонение в течение 5 лет после месяца вашего рождения. Отобразить данные последнего года на графике. Результаты сохранить в виде блокнота Jupyter. 


```python
import pandas as pd
df1 = pd.read_csv('https://datahub.io/core/oil-prices/r/brent-month.csv')
df1.head()
```

```python
mydf = df1[(df1['Date']>='2001-04-01') & (df1['Date']<'2006-04-01')]
print('Средняя цена:\t\t{0:.4}'.format(mydf['Price'].mean()))
print('Минимальная цена:\t{0}'.format(mydf['Price'].min()))
print('Максимальная цена:\t{0}'.format(mydf['Price'].max()))
print('СКО цены:\t\t{0:.4}'.format(mydf['Price'].std()))
```

```python
import matplotlib.pyplot as plt
mydf_lastyear = mydf[mydf['Date'] >= '2005-04-01']
plt.plot(mydf_lastyear['Date'], mydf_lastyear['Price']) 
plt.xticks(rotation=90)
plt.show()
```
